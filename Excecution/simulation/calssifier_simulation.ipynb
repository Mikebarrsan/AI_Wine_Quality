```python
from pathlib import Path

import numpy as np
import pandas as pd

from tensorflow.keras.models import load_model

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
model = load_model("../../models/1D_CNN_checkpoints/best_1D_CNN_model.keras")

model.summary()
```

```python
attributes = [
    "fixed acidity",
    "volatile acidity",
    "citric acid",
    "residual sugar",
    "chlorides",
    "free sulfur dioxide",
    "total sulfur dioxide",
    "density",
    "pH",
    "sulphates",
    "alcohol"
]

def ask_float(attribute_name):
    while True:
        value = input(f"{attribute_name}: ").strip()

        if value == "":
            print("Value cannot be empty. Please enter a number.")
            continue

        try:
            print(attribute_name , ": ", value)
            return float(value)
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

def insert_wine_data():
    wine_data = {}

    print("Please enter the wine data:\n")

    for attribute in attributes:
        wine_data[attribute] = ask_float(attribute)

    single_wine_df = pd.DataFrame([wine_data])

    return single_wine_df
```

```python
single_wine_df = insert_wine_data()

quality = int(input("Quality (0-10): ").strip())
```
